# Image Translaion


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread(r'C:\Users\USER\Documents\csc418\week_2\img\puppy.jpeg', 0)
rows, cols = img.shape

#Plot the original image 
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

M = np.float32([[1, 0, 100], [0, 1, 50]]) #(x,y)
transl = cv.warpAffine(img, M, (cols, rows))

#Plot the translated image
plt.subplot(1,2,2)
plt.title("Translated Image")
plt.imshow(transl)

cv.waitKey(0)
cv.destroyAllWindows()
```
