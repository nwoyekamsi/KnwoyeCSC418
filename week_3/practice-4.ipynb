# IMAGE CROPPING

```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2 as cv
img = cv.imread(r'C:\Users\USER\Documents\csc418\week_2\img\puppy.jpeg', 0)

#plot the original image 
plt.subplot(1,2,1)
plt.title("original")
plt.imshow(img)

cropped_img = img[50:200, 100:300]

#Plot the original image
plt.subplot(1,2,2)
plt.title("Cropped Image")
plt.imshow(cropped_img)

cv.imwrite('cropped_out.jpg', cropped_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
