# REMOVING NOISE


```python
#Import the necessary libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

#Load the image
image = cv2.imread(r'img\puppy.jpeg')

#plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

#remove noise using a medium filter
filtered_image = cv2.medianBlur(image, 15)

#Save the image 
cv2.imwrite('img/Medium-Blur.jpg', filtered_image)

#plot the blureded image
plt.subplot(1,2,2)
plt.title("Median blur")
plt. imshow(filtered_image)
plt.show()
```
